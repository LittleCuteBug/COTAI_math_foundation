```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
#1.6
fig = plt.figure()
ax = plt.axes(projection='3d')
# data
t = np.linspace(-1,2,10)
k = np.linspace(0,1,10)
v = np.array([2,3,4])
u = np.array([3,4,5])
w_x = t*u[0] + v[0]
w_y = t*u[1] + v[1]
w_z = t*u[2] + v[2]
ax.quiver3D([0],[0],[0],[u[0]],[u[1]],[u[2]], color = 'red') #vector u
ax.scatter3D([v[0]],[v[1]],[v[2]],color = 'blue') # point v
ax.plot3D(w_x, w_y, w_z, color = 'green') # line L1 
plt.show()
```

```python
#1.9
fig = plt.figure()
ax = plt.axes(projection='3d')
# data
t = np.linspace(-1,2,10)
k = np.linspace(0,1,10)
p = np.array([5,2,3])
u = np.array([3,4,5])
w_x_0 = t*u[0] 
w_y_0 = t*u[1] 
w_z_0 = t*u[2]
w_x_1 = t*u[0] + p[0] 
w_y_1 = t*u[1] + p[1]
w_z_1 = t*u[2] + p[2]
ax.quiver3D([0],[0],[0],[p[0]],[p[1]],[p[2]], color = 'blue') #vector u
ax.plot3D(w_x_0, w_y_0, w_z_0, color = 'green') # line L0 
ax.plot3D(w_x_1, w_y_1, w_z_1, color = 'red') # line L2 
ax.scatter3D([0],[0],[0],color = 'orange') # zero vector
ax.scatter3D([p[0]],[p[1]],[p[2]],color = 'yellow') # point v
plt.show()
```
